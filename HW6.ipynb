```python
# Dependencies
import matplotlib.pyplot as plt
import requests
import json
import pandas as pd
import numpy as np
from My_config import api_key
import random 
from citipy import citipy
import time
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
# generate the list of random cities using latitude and longitude
lat_lngs = []
cities = []
countries=[]
lat=[]
lng=[]

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1600)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1600)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
i=0
j=0
for lat_lng in lat_lngs:
    i+=1
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    country = citipy.nearest_city(lat_lng[0], lat_lng[1]).country_code
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        countries.append(country)
        lat.append(lat_lng[0])
        lng.append(lat_lng[1])
    else:
        print(f"This city No.{i} is already in the list.  Skipping...")
        j+=1
print('-------------------------------------------------')
print(f'Total {j} citites were skipped.\n')
```

```python
dict = {
    'city': cities,
    'country':countries,
    'lat' : lat,
    'lng' : lng,
}

df=pd.DataFrame(dict)
#df
```

```python
# Loop through the list of cities and perform a API request for data on each
print(f"\nBeginning Data Retrieval") 
print('----------------------------------')

j=0
for index,row in df.iterrows():
    try:                 
        time.sleep(1)
        city=row["city"]   
        print(f"Processing Record {index+1} | {city}") 
        info= requests.get(query_url + city).json()
        df.loc[index, "date"] = format(info['dt'], '10')
        df.loc[index, "max_temp"] = info['main']['temp_max']
        df.loc[index, "humidity"] =info['main']['humidity']
        df.loc[index, "cloudiness"]=info['clouds']['all']
        df.loc[index, "windspeed"]=info['wind']['speed'] 
    except (KeyError, IndexError):
        j+=1
        print(f"City not found in weather API.  Skipping...")   
        
print('------------------------------------------------')
print(f"Data Retrieval Complete ") 
print(f'Total {j} citites were not found(skipped) through weather API.')
print('------------------------------------------------')
#df
```

```python
#drop the NaN rows
df=df.dropna()
df.count()
```

```python
# Export the city data into a cities.csv
df=df.reset_index(drop=True)
df.to_csv("cities.csv")
df.head()
```

```python
#convert the timestamp to datetime format(Month/Day/Year)
df['date']=pd.to_datetime(df['date'] , unit='s',errors='coerce')
df['date']=df['date'].dt.strftime('%m/%d/%Y')
df['date']
day=df.iloc[0,4]
```

```python
%matplotlib notebook
plt.scatter(df["lat"], df["max_temp"], marker="o",edgecolors="black", alpha=0.7)

# Incorporate the other graph properties
plt.title(f"City Latitude vs Max Temperature ({day})")
plt.ylabel(f"Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Show plot
plt.show()
plt.savefig(f"./max_temp.png")
```
